```python
import pandas as pd
```

# Divide into train and test data

```python
data = pd.read_csv("./data/train.csv")
```

```python
y = data["rainfall"]
X = data[["id","pressure","maxtemp","temparature","mintemp","dewpoint","humidity","cloud","sunshine","winddirection","windspeed"]]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1)
```

# Check for MC, Distributions

```python
X_train = X_train.set_index("id")
X_test = X_test.set_index("id")
```

```python
X_train.corr()
```

```python
# keep only temperature, pressure, humidity, cloud, sunshine, windspeed
```

```python
X_train = X_train[["temparature","pressure","humidity","cloud","sunshine","windspeed"]]
X_test = X_test[["temparature","pressure","humidity","cloud","sunshine","windspeed"]]
```

# Logsitic Regression

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
clf = LogisticRegression(random_state=0,max_iter = 500).fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.feature_names_in_
```

```python
test = pd.read_csv("./data/test.csv")
```

```python
X_predict = test.set_index("id")
```

```python
X_predict
```

```python
X_predict = X_predict[["temparature","pressure","humidity","cloud","sunshine","windspeed"]]
```

```python
y_predict = clf.predict(X_predict)
```

```python
df = pd.DataFrame(columns = ["id","rainfall"])
```

```python
df["id"] = X_predict.index
```

```python
df["rainfall"] = y_predict
```

```python
df = df.set_index("id")
```

```python
df.to_csv("./data/submissions_2.csv")
```
